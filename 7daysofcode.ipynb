Criando a coluna de acordo com a Classificação Decimal Universal abaixo:

000 a 099: Generalidades. Ciência e conhecimento.

100 a 199: Filosofia e psicologia.

200 a 299: Religião.

300 a 399: Ciências sociais.

400 a 499: Classe vaga. Provisoriamente não ocupada.

500 a 599: Matemática e ciências naturais.

600 a 699: Ciências aplicadas.

700 a 799: Belas artes.

800 a 899: Linguagem, língua, linguística.

900 a 999: Geografia, biografia, história.

No campo "localização", temos um código numérico referente a essa classificação. Portanto, iremos criar uma nova coluna que refletirá a localização por extenso, de acordo com a classificação.



```python
cdu_lista = []
for cdu in emprestimos['localizacao']:
    if cdu < 100:
        cdu_lista.append('Generalidades')
    elif cdu < 200:
        cdu_lista.append('Filosofia e Psicologia')
    elif cdu < 300:
        cdu_lista.append('Religiao')
    elif cdu < 400:
        cdu_lista.append('Ciências sociais.')
    elif cdu < 500:
        cdu_lista.append('Classe vaga')
    elif cdu < 600:
        cdu_lista.append('Matemática e ciências naturais')
    elif cdu < 700:
        cdu_lista.append('Ciências aplicadas')
    elif cdu < 800:
        cdu_lista.append('Belas artes')
    elif cdu < 900:
        cdu_lista.append('Linguagem, língua, linguística')
    else:
        cdu_lista.append('Geografia, biografia, história')
    
emprestimos['CDU'] = cdu_lista
```

A diretoria também gostaria de gerenciar melhor os recursos humanos da biblioteca de acordo com a demanda de trabalho existente. Por exemplo:


gerenciar a programação de férias dos colaboradores de acordo com os meses de menor demanda;

Há uma suspeita interna de que os meses com maior número de exemplares emprestados sejam março e setembro, mas não foi realizada uma análise real sobre isso.


```python
emprestimos['mes_emprestimo'] = emprestimos['data_emprestimo'].dt.month 
```

```python
quantidade_exemplares_mes = emprestimos.groupby('mes_emprestimo')['id_exemplar'].count()
```

```python
quantidade_exemplares_mes.plot(
    kind='line',
    xlabel='Mês',
    ylabel='Quantidade de Exemplares Emprestados',
    title='Empréstimos por Mês',
    figsize=(20, 4)
)
```

## Análise do Gráfico de Empréstimos por Mês

Ao analisar o gráfico de empréstimos mensais, é possível identificar claramente os períodos de maior e menor demanda na biblioteca ao longo do ano. Os meses com menor quantidade de exemplares emprestados representam oportunidades ideais para o gerenciamento interno da equipe.

**Sugestões para a Programação de Férias e Atividades Internas:**

- **Programação de Férias:**  
  Recomenda-se que as férias dos colaboradores sejam preferencialmente agendadas para os meses de menor demanda(6,7), conforme indicado pelo gráfico. Dessa forma, a ausência de funcionários terá menor impacto no atendimento ao usuário.

- **Atividades Não Relacionadas ao Atendimento:**  
  Atividades como inventário, organização do acervo, treinamentos internos e manutenção podem ser planejadas para esses mesmos períodos de baixa demanda. Assim, é possível otimizar o tempo da equipe sem prejudicar o serviço ao público.

**Conclusão:**  
O uso do gráfico permite uma gestão mais eficiente dos recursos humanos, garantindo que o atendimento ao usuário não seja comprometido nos períodos de maior movimento e que as atividades internas sejam realizadas de forma estratégica nos meses mais tranquilos.


Além do gerenciamento anual das atividades, a diretoria também necessita que seja planejada uma programação diária das atividades. Por este motivo, verifique quais foram os horários com maior quantidade de empréstimos ao longo de um dia inteiro.


```python
emprestimos['hora_emprestimo'] = emprestimos['data_emprestimo'].dt.hour 
```

```python
quantidade_exemplares_hora = emprestimos.groupby('hora_emprestimo')['id_exemplar'].count()
quantidade_exemplares_hora.plot(
    kind='line',
    xlabel='Hora',
    ylabel='Quantidade de Exemplares Emprestados',
    title='Empréstimos por Hora',
    figsize=(10, 5)
)
```

## Análise dos Empréstimos por Hora

O gráfico de empréstimos por hora mostra claramente que a demanda na biblioteca é significativamente maior entre 8h e 18h, com picos notáveis por volta das 10h, 15h e 17h. Fora desse intervalo, especialmente antes das 8h e após as 20h, o número de empréstimos é muito baixo ou praticamente inexistente.

**Horários de Maior Movimento:**
- Entre 8h e 18h, com destaque para os picos próximos das 10h, 15h e 17h.

**Horários de Menor Movimento:**
- proximo das 8h e das 20h, a quantidade de empréstimos é mínima.

### Recomendações para Alocação de Atividades Internas

- **Atividades que não envolvem atendimento ao usuário** (como inventário, organização do acervo, treinamentos e manutenção) devem ser preferencialmente realizadas nos horários de menor movimento, ou seja, proximo as 8h e 20h.
- Caso não seja possível realizar essas atividades fora do horário de pico, recomenda-se priorizar os intervalos entre 12h e 14h, quando há uma leve redução na demanda.
